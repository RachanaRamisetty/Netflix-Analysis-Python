# Analysis of Quarterly Data For Competitors Based on Revenue and Subscribers

```python
#importing Numpy,Panda,Matplotlib and Seaborn libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
os.listdir(os.getcwd()+"/NetflixData/Content")
```

```python
Revenue_Quarterly = pd.read_csv("NetflixData/Competition/Revenue_Quarterly.csv")
Revenue_Quarterly #Accessing the Quarterly Revenue data
```

```python
Subscribers_Quarterly = pd.read_csv("NetflixData/Competition/Quarterly_Subscribers.csv")
Subscribers_Quarterly #Accessing the Quarterly Subscriber Data
```

```python
pie_chart_plotdata = Revenue_Quarterly.set_index('Quarter') #Setting the column Quarter to the Index
```

```python
#Plotting a pie chart for the Q1 2018
explode = (0.1,0,0)
colors = ['Red','slategrey','lightgreen']
pie_chart_plotdata.loc["Q1 2018"].plot(kind='pie',explode = explode,colors= colors, autopct='%.1f%%',textprops={'fontsize': 14})

sns.set(rc = {'figure.figsize':(10,10)})
plt.ylabel('')
plt.xlabel('')
plt.title('Market Share by Value for Q1 2018',fontsize ='18',fontweight ='bold')
plt.savefig('i21.png')
plt.show()
```
